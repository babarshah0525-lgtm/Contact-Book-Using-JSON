```python
import json

# The 'contacts' variable must be loaded *after* the load_contacts function is defined,
# so we load it inside main().

def load_contacts():
    """Loads contacts from 'file.json' or returns an empty dictionary if the file doesn't exist."""
    try:
        with open("file.json", "r") as file:
            contacts_data = json.load(file)
            return contacts_data # FIX 1: Must return the data if the file is found
    except FileNotFoundError:
        contacts_data = {}
        return contacts_data
    # Note: In your original code, 'contacts' was used in the try block and 'contacts_data'
    # in the except block. I've standardized to 'contacts_data' in the except block and 
    # 'contacts' in the try block to match your original variable names.

def save_contacts(contacts):
    """Saves the current contacts dictionary to 'file.json'."""
    # FIX 2: Removed the extraneous backtick (`) at the end of the line
    with open("file.json", "w") as file:
        json.dump(contacts, file, indent=4)

def add_contacts(contacts):
    """Prompts the user for contact details and adds them to the contacts dictionary, then saves."""
    name = input("Enter Name of Contact: ")
    number = input("Enter Number of User: ")
    email = input("Enter Email of User: ")
    
    contacts[name] = {"phone": number, "Email": email}
    
    # FIX 3: Must save the contacts after adding a new one
    save_contacts(contacts)
    print(f"\nContact '{name}' added and saved.")
    # Removed print(contacts) for cleaner output

def delete_contacts(contacts):
    """Prompts the user for a contact name and deletes it if found."""
    name = input("Enter Name of Contact to Delete: ")
    
    if name in contacts:
        del contacts[name]
        save_contacts(contacts)
        print(f"\nContact '{name}' deleted and file updated.")
    else:
        print(f"\nError: Contact '{name}' not found.")

def view_contacts(contacts):
    """Prints all contacts in a formatted list."""
    if not contacts:
        print("\nNo contacts found.")
        return # Exit the function if there are no contacts
        
    print("\n--- Contact List ---")
    for name, details in contacts.items():
        print(f"Name : {name}")
        # FIX 4: Dictionary keys must be accessed using string literals ('phone', 'Email')
        print(f"Phone Number : {details['phone']}") 
        print(f"Email : {details['Email']}") # Note: Capital 'E' for 'Email' to match add_contacts
        print("-" * 25)
    print("--------------------")

# ---

def main():
    # FIX 5: Contacts are loaded here, after the function is defined.
    contacts = load_contacts() 
    
    print("Welcome to the CONTACT BOOK")
    while True:
        try:
            operation = int(input("Enter 1. to ADD Contact\n2. Delete Contact\n3. View Contacts\n4. Exit\nYour Choice: "))
            
            if operation == 1:
                # FIX 6: Removed print() around the function call to prevent printing 'None'
                add_contacts(contacts)
            elif operation == 2:
                # FIX 6: Removed print()
                delete_contacts(contacts)
            elif operation == 3:
                # FIX 6: Removed print()
                view_contacts(contacts)
            elif operation == 4:
                print("Thank you for using the CONTACT BOOK. Goodbye! ðŸ‘‹")
                break
            else:
                print("\nThe INPUT is Invalid! Please Try Again (1-4).")
        except ValueError:
            print("\nInvalid input! Please enter a number (1, 2, 3, or 4).")
        except Exception as e:
            print(f"\nAn unexpected error occurred: {e}")

if __name__=="__main__":
    main()
```
